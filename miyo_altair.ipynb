```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('miyo_data_frame.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.head()
```

```python
alt.Chart().mark_point(filled=True, size=5, opacity=0.5).encode(
  alt.X(alt.repeat('column'), type='quantitative'),
  alt.Y(alt.repeat('row'), type='quantitative')
).properties(
  width=100,
  height=100
).repeat(
  data=df,
  row=['solarVoltage'],
  column=['brightness', 'frequency', 'moisture', 'rssi', 'temperature', 'time_hour']
)
```

```python
alt.Chart().mark_point(filled=True, size=10, opacity=0.5).encode(
  alt.X(alt.repeat('column'), type='temporal'),
  alt.Y(alt.repeat('row'), type='quantitative'),
  tooltip = [
      alt.Tooltip(alt.repeat("row"), type="quantitative"),
      alt.Tooltip(alt.repeat("column"), type="temporal")
  ]
).properties(
  width=800,
  height=100
).repeat(
  data=df,
  row=['brightness', 'frequency', 'moisture', 'rssi', 'temperature', 'time_hour'],
  column=['timestamp']
)
```
